```python
import numpy as np
import photonforge as pf

# Set the default project technology
tech = pf.basic_technology()
pf.config.default_technology = tech

s_bend = pf.parametric.s_bend(port_spec="Strip", length=5, offset=1.5, model=pf.Tidy3DModel())
# Length -> Horizontal length of the S-bend
# Offset -> Vertical offset between the two ports

# Get the S parameters (uses the default Tidy3D Model)
wavelengths = np.linspace(1.5, 1.6, 200)
freqs = pf.C_0 / wavelengths
# convert wavelenghths, pf.C_0 is the speed of light in um/ps and wavelength is in um
s_matrix = s_bend.s_matrix(freqs)

# We can easily plot the S parameters with the following utility
fig, ax = pf.plot_s_matrix(s_matrix)

# since the EM simulation operate in frequenyc space, we convert wavelengths to freqs
```
